```python
### BMI calculator ###

height = 1.70
weight = 65

bmi = weight / (height**2)

print ("The BMI of " + str(weight) +"kg and "+str(height)+"m is " + str(bmi)+".")
```

```python
### Score calculation ###

grade = 90

if grade>=90:
    print ("A")
elif grade>=70:
    print ("B")
else:
    print ("F")
```

```python
### Blood donation calculator ###

donor = 'A' #('AB'/'A'/'B','O').
recipient =  'O'  #('AB'/'A'/'B','O').

if donor != 'O' and donor != 'A' and donor != 'B' and donor != 'AB':
    print ('Invalid input')

elif recipient != 'O' and recipient != 'A' and recipient != 'B' and recipient != 'AB':
    print ('Invalid input')

elif donor == 'O' or recipient == 'AB':
    print (True)
else:
    if donor == recipient:
        print (True)
    else:
        print (False)
```
