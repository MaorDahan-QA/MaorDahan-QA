```python
### The number of even elements in the matrix ###

mat = [[1,4,5],[2,3,4]]

count =0

for row in mat:
    for n in row:
        if n%2==0:
            count=count+1


for i in range(len(mat)):
    for j in range(len(mat[0])):
        if mat[i][j]%2==0:
            count=count+1

print(count)
```

```python
### Add 10 points to every grade below 60 and average of the updated matrix ###

grades = [[40, 80, 90], [60, 90, 95], [17, 20,100]]

for i in range(len(grades)):
    for j in range(len(grades[0])):
        if grades[i][j]<60:
            grades[i][j]=grades[i][j]+10

print (grades)

sm=0
count=0
for i in range(len(grades)):
    for j in range(len(grades[0])):
        sm=sm+grades[i][j]
        count=count+1

print (sm/count)
```

```python
### Square and Pyramid of star ###

n = 5

## prints a square of n*n stars ##
for i in range (1,n+1):
    for j in range (1,n+1):
        print ('*',end='')
    print()

## prints stars pyramid ##
for i in range (1,n+1):
    for j in range (1,i+1):
        print ('*',end='')
    print()
```

```python
### Sort numbers in the list ###

lst = [5, 2, 7, 9, 8, 4]

print ("Even numbers:")

## for loop ##
for num in lst:
    if num%2==0:
        print(num)

print ("\n")

## while loop ##
i = 0
while i<len(lst):
    if lst[i]%2==0:
        print(lst[i])
    i=i+1

print ("\n\nOdd numbers:")

## for loop ##
for i in range(len(lst)):
    if i%2==0:
        print (lst[i])

print ("\n")

## while loop ##
i = 0
while i<len(lst):
    if i%2==0:
        print(lst[i])
    i=i+1
```
