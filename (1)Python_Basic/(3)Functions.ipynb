```python
### A function that returns a string of capital letters ###

def get_caps(s):
    s2=''
    for c in s:
        if c.isupper():
            s2=s2+c
    return s2

print (get_caps('Hello Friend'))
```

```python
### A function that returns the name of the website with the highest traffic ###

def get_max_index(lst):
   mxi=0
   for i in range(len(lst)):
       if lst[i]>lst[mxi]:
           mxi=i
   return mxi

def get_max_traffic(websites, traffic):
    mxi = get_max_index(traffic)
    return websites[mxi]

websites = ['Ynet','NRG','Haaretz','Mako','CNN','Walla']
traffic = [100,10,80,50,600,80]

print (get_max_traffic(websites, traffic))
```
