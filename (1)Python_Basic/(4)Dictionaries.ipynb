```python
### Sorted letters in a string ###

text = 'supercalifragilisticexpialidocious'

def get_char_count(text):
    char_count = {}
    for char in text:
        count = char_count.get(char, 0)
        count += 1
        char_count[char] = count
    return char_count


def print_by_keys_order(char_count):
    chars = char_count.keys()
    sorted_chars = sorted(chars)
    for char in sorted_chars:
        print(char, ':', char_count[char])


def print_by_values_order(char_count):
    chars = char_count.keys()
    sorted_chars = sorted(chars, key=char_count.get, reverse=True)
    for char in sorted_chars:
        print(char, ':', char_count[char])


char_count = get_char_count(text)
print("-------- keys sorted alphabetically -------")
print_by_keys_order(char_count)
print("-------- keys sorted by values -------")
print_by_values_order(char_count)
```

```python
### Function returns a list containing the two most frequenct words in a given text ###

def find_most_freq_words_in_text(text):
    words = text.split()
    d= {}
    for word in words:
        if word not in d:
            d[word]=1
        else:
            d[word]=d[word]+1
    sorted_words = sorted(d.keys(),key=d.get, reverse=True)
    return sorted_words[:2]

text = "Hello my name is Adi what is your name is your name Nelly?"
print (find_most_freq_words_in_text(text))
```

```python
### Return a dictionary mapping the number of launches for each city ###

def calc_launches_freq(launches):
    d = {}
    for launch in launches:
        if launch[0] not in d:
            d[launch[0]]=1
        else:
            d[launch[0]]=d[launch[0]]+1
    return d

## Print a list of launches, sorted by city (sort by keys) ##
def print_dic_by_keys(d):
    cities = d.keys()
    sorted_cities = sorted(cities)
    for city in sorted_cities:
        print (city,':',d[city])

## Print a list of launches, sorted by number of launches (sort by values) ##
def print_dic_by_values(d):
    cities = d.keys()
    sorted_cities = sorted(cities,key=d.get,reverse=True)
    for city in sorted_cities:
        print (city,':',d[city])

## Function receives a dictionary mapping the number of launches for each city ##
def convert_dic(d):
    d2={}
    for city in d.keys():
        num = d[city]
        if num not in d2:
            d2[num]=[city]
        else:
            d2[num].append(city)
    return d2

Launches = [('Sderot','07:30'), ('Nahal-Oz','09:01'),('Sderot','09:02'),('Ofakim','10:30'),('Sderot','11:00')]

dic = calc_launches_freq(Launches)
print_dic_by_keys(dic)
print ('-------')
print_dic_by_values(dic)
print (convert_dic(dic))
```
