```
import tweepy

ckey="***" # Add the ckey instead of the ***
csecret="***" # Add the csecret instead of the ***
atoken="***" # Add the atoken instead of the ***
asecret="***" # Add the asecret instead of the ***

OAUTH_KEYS = {'consumer_key':ckey, 'consumer_secret':csecret,'access_token_key':atoken, 'access_token_secret':asecret}
auth = tweepy.OAuthHandler(OAUTH_KEYS['consumer_key'], OAUTH_KEYS['consumer_secret'])
api = tweepy.API(auth)

words = ["#DonaldTrump OR #MAGA OR #Republican OR #Trump OR #Trump2020 OR #Trump2024 OR #TrumpWon OR DonaldTrump OR MAGA OR Republican OR Trump OR Trump2020 OR Trump2024 OR TrumpWon"]

Tweets = tweepy.Cursor(api.search, q=words, exclude= "retweets", lang="en", tweet_mode='extended', since="2021-01-20").items(10000)

for tweet in Tweets:
    print(tweet.user.location, ";", tweet.created_at, ";", tweet.user.screen_name, ";", tweet.full_text.replace('\n', '').replace('\"', '').replace('\r', ''))
```
