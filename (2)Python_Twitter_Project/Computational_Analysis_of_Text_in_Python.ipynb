```python
filenameAFINN = 'AFINN-111.txt'
afinn = pd.read_csv(filenameAFINN,
                          header=None, sep='\t', names=['term', 'value'])
afinn = afinn.set_index('term')
afinn = afinn['value'].to_dict()
```

```python
afinn
```

```python
pattern_split = re.compile(r"\W+")
```

```python
def sentiment(text):
    words = pattern_split.split(text.lower())
    sentiments = [afinn.get(word, 0) for word in words]
    if sentiments:
        sentiment = float(sum(sentiments))/math.sqrt(len(sentiments))

    else:
        sentiment = 0
    return sentiment
```

```python
lst_analysis_Tweets = []
for i in Text_for_analysis:
  lst_analysis_Tweets.append(sentiment(i))
```

```python
print (lst_analysis_Tweets)
print ("average: ", sum(lst_analysis_Tweets)/len(lst_analysis_Tweets))
```
