```python
#@title don't show code till dblclick

# פונקציה שבודקת את הקלט
def check_valid_input(letter_guessed, old_letters_guessed):
  global special_characters, list_letters, word_len, n, word, letter

  if len(letter_guessed)>1 or any(x in special_characters for x in letter_guessed) or letter_guessed in old_letters_guessed:
    print ("X\n")
    print (*list_letters, sep=' -> ')
    print ("\n")
    print (*word_len, sep=' ')
    print ("\n")
    print_hangman(n)
    return

  else:
    list_letters.append(letter)
    list_letters.sort()
    show_hidden_word (word, list_letters)
    print_hangman(n)
    return


# פונקציה שבודקת האם הקלט נמצא במילה
def show_hidden_word(secret_word, old_letters_guessed):
  i = 0
  global word_len, word, letter

  while i < len(secret_word):
    if secret_word[i] in old_letters_guessed:
      word_len[i] = secret_word[i]
    i=i+1

  print ("\n")
  print (*word_len, sep=' ')
  print ("\n")
  check_win(word, letter)

  return


# פונקציה שבודקת האם המשתמש ניצח
def check_win(secret_word, letter):
  global word_len, n

  if "_" in word_len and letter not in secret_word:
    n = n + 1

  elif "_" not in word_len:
    print ("You Win")

  return


# פונקציה שמדפיסה את האיש התלוי
def print_hangman(num_of_tries):
  global pictures

  if num_of_tries < 7:
    print (pictures[num_of_tries], "\n")

  else:
    print (pictures[7], "\n")
    print ("Game Over", "\n")

  return


# פונקציה שבוחרת מילה
def choose_word(list):
  global word, word_len

  a = 0
  j = len(list)
  location_word = ra.randint(0, j-1)

  word = list[location_word].lower()

  word_len = "_"*len(word)
  word_len = [x for x in word_len]

  for x in word:
    if x == " ":
      word_len[a] = " "
    a = a+1

  print (*word_len, sep=' ')

  return
```
